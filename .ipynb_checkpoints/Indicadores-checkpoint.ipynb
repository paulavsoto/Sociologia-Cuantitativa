**Indicadores Relevantes**

**Introduction**


The analysis period spans from 1990 to 2024 on a global scale, but focuses on countries that allow for relevant comparisons, particularly to situate Chile within the Latin American context.

The aim is to contrast opinions with objective data on Chile, especially when it is described as a leading country in the region. Likewise, it examines the demands for greater state funding—based on the argument that Chile has sufficient resources—versus those who claim that resources are limited and advocate for private sector involvement to meet social demands.

```python
import seaborn as sns
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd
from pandas_datareader import data, wb
```

**Bajando datos del Banco Mundial**

```python
#Obtener un listado de indicadores, en este caso económicos, 
#definidos por la sigla GDP, PRODUCTO INTERNO EN BRUTO en inglés
wb.search('gdp')
```

```python
# Obtengamos la lista de paises disponibles
countries=wb.get_countries()
countries.head()
```

**SI.POV.GINI Indice de GINI**

```python
# Definir el indicador y el rango de fechas
indicador = 'SI.POV.GINI'
inicio = '1990'
fin = '2024'

# Lista de países a considerar (usando códigos ISO3C)
paises_seleccionados = ['CHE', 'NOR', 'ISL', 'CHN', 'JPN', 'DEU', 'ARG', 'CHL', 'BRA', 'MEX', 'PRT']

# Descargar datos solo para los países seleccionados
global_gini = wb.download(
    indicator=indicador,
    country=paises_seleccionados,  # Solo los países de la lista
    start=inicio,
    end=fin
)

global_gini.reset_index(inplace=True)

# Imprimir información del DataFrame
global_gini.info()
```

```python
plt.figure(figsize=(22, 16))
sns.set_style("whitegrid")

# Definir variables necesarias
unique_countries = global_gini['country'].unique()
all_years = sorted(global_gini['year'].unique())
even_years = all_years[::2]  # Selecciona años pares
palette = sns.color_palette("husl", len(unique_countries))
markers = ['o', 's', 'D', '^', 'v', '<', '>', 'p', '*', 'h', 'H', '+', 'x', 'd']

# Filtrar datos para el período 2019-2023
df_filtrado = global_gini[global_gini['year'].astype(int).between(2019, 2023)]

# Crear una figura con 4 subplots (2 filas x 2 columnas)
fig, axs = plt.subplots(2, 2, figsize=(22, 16))
plt.subplots_adjust(hspace=0.4, wspace=0.2, bottom=0.1)

# Filtrar datos solo para Chile
df_chile = global_gini[global_gini['country'] == 'Chile']

# Primer gráfico superior izquierdo - Todos los países (1990-2024)
for i, country in enumerate(unique_countries):
    country_data = global_gini[global_gini['country'] == country]
    axs[0, 0].plot(
        country_data['year'],
        country_data['SI.POV.GINI'],
        label=country,
        color=palette[i],
        marker=markers[i % len(markers)],
        linewidth=2.5,
        markersize=8
    )
axs[0, 0].set_xticks(even_years)
axs[0, 0].set_xticklabels(even_years, rotation=45)
axs[0, 0].invert_xaxis()
axs[0, 0].set_title('Evolución del Coeficiente GINI - Todos los países (1990-2024)', fontsize=16)
axs[0, 0].set_xlabel('Año', fontsize=14)
axs[0, 0].set_ylabel('Coeficiente GINI', fontsize=14)
axs[0, 0].grid(True, linestyle='--', alpha=0.7)

# Segundo gráfico superior derecho - Todos los países (2019-2023)
for i, country in enumerate(unique_countries):
    country_data = df_filtrado[df_filtrado['country'] == country]
    axs[0, 1].plot(
        country_data['year'],
        country_data['SI.POV.GINI'],
        label=country,
        color=palette[i],
        marker=markers[i % len(markers)],
        linewidth=2.5,
        markersize=8
    )
axs[0, 1].invert_xaxis()
axs[0, 1].set_title('Evolución del Coeficiente GINI - Todos los países (2019-2023)', fontsize=16)
axs[0, 1].set_xlabel('Año', fontsize=14)
axs[0, 1].grid(True, linestyle='--', alpha=0.7)

# Tercer gráfico inferior izquierdo - Chile todos los años
axs[1, 0].plot(
    df_chile['year'],
    df_chile['SI.POV.GINI'],
    label='Chile',
    color='#1f77b4',
    marker='o',
    linewidth=3,
    markersize=10
)
axs[1, 0].set_xticks(even_years)
axs[1, 0].set_xticklabels(even_years, rotation=45)
axs[1, 0].invert_xaxis()
axs[1, 0].set_title('Evolución del Coeficiente GINI en Chile (1990-2024)', fontsize=16)
axs[1, 0].set_xlabel('Año', fontsize=14)
axs[1, 0].grid(True, linestyle='--', alpha=0.7)

# Cuarto gráfico inferior derecho - Chile (2018-2024)
df_chile_reciente = df_chile[df_chile['year'].astype(int) >= 2018]
axs[1, 1].plot(
    df_chile_reciente['year'],
    df_chile_reciente['SI.POV.GINI'],
    label='Chile',
    color='#d62728',
    marker='s',
    linewidth=3,
    markersize=10
)
axs[1, 1].invert_xaxis()
axs[1, 1].set_title('Evolución del Coeficiente GINI en Chile (2018-2024)', fontsize=16)
axs[1, 1].set_xlabel('Año', fontsize=14)
axs[1, 1].grid(True, linestyle='--', alpha=0.7)

# Leyenda
handles, labels = axs[0, 0].get_legend_handles_labels()
fig.legend(
    handles=handles,
    labels=labels,
    title='Países',
    loc='lower center',
    bbox_to_anchor=(0.5, -0.05),
    fontsize=12,
    title_fontsize='14',
    ncol=min(6, len(unique_countries)))

plt.tight_layout()
plt.show()
```
